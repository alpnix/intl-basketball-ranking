```python
import pandas as pd
```

## Read in the Data

```python
acb = pd.read_csv('ligaacb-2324.csv')
euro = pd.read_csv('eurocup-2324.csv')

acb = acb[['Date', 'Team', 'PTS', 'Opp', 'PTS.1']]
euro = euro[['Date', 'Team', 'PTS', 'Opp', 'PTS.1']]
acb = acb.rename(columns={'PTS.1': 'PTS_Opp'})
euro = euro.rename(columns={'PTS.1': 'PTS_Opp'})

acb
```

## Weight the Data

```python
acb["Weight"] = 0.5
euro["Weight"] = 1
```

## Merge the data

```python
df = pd.concat([acb, euro])

df['Date'] = pd.to_datetime(df['Date'], format='%a %b %d %Y')
df = df.sort_values(by=['Date'])
df.reset_index(drop=True, inplace=True)

df
```

## Save the Data

```python
df.to_csv('ligaacb-eurocup-2324.csv', index=False)
```
